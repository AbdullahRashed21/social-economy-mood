```python
vol=dfs.drop(columns=['label','score'])
vol=dfs.groupby(['Tweet_time']).count()
```

```python
vol.columns = vol.columns.str.replace('label', 'number_of_tweets')
```

```python
vol
```

```python

dfsp=dfs[dfs["label"].str.contains("positive")]
dfsp.columns = dfsp.columns.str.replace('label', 'pos')
dfsp.columns = dfsp.columns.str.replace('score', 'pos_score')
dfsp=dfsp.groupby(pd.Grouper(key='Tweet_time', axis=0,freq='D', sort=False)).sum()
dfsp=dfsp.merge(vol, how='inner', on='Tweet_time')
dfsp['Pos_weighted_average']=(dfsp.pos_score/dfsp.number_of_tweets)
dfsp=dfsp.drop(columns=['number_of_tweets', 'pos_score'])
dfsp
```

```python
dfsn=dfs[dfs["label"].str.contains("negative")]
dfsn.columns = dfsn.columns.str.replace('label', 'neg')
dfsn.columns = dfsn.columns.str.replace('score', 'neg_score')
dfsn=dfsn.groupby(pd.Grouper(key='Tweet_time', axis=0,freq='D', sort=False)).sum()
dfsn=dfsn.merge(vol, how='inner', on='Tweet_time')
dfsn['neg_weighted_average']=(dfsn.neg_score/dfsn.number_of_tweets)
dfsn=dfsn.drop(columns=['number_of_tweets', 'neg_score'])
dfsn
```

```python
fd=dfsp.merge(dfsn, how='inner', on='Tweet_time')
```

```python
#dfs['i']=(np.sqrt(dfs.pos*dfs.pos+dfs.neg*dfs.neg))
fd['i']=(np.sqrt(fd.Pos_weighted_average*fd.Pos_weighted_average+fd.neg_weighted_average*fd.neg_weighted_average))
#dfs['angle']=(2*dfs.i+2*dfs.pos-2*dfs.i*dfs.pos* np.cos(dfs.neg))
fd['theta']=(2*fd.i+2*fd.Pos_weighted_average-2*fd.i*fd.Pos_weighted_average*np.cos(fd.neg_weighted_average))
#dfs['Polarity']=(1-4*dfs.angle/np.pi)
fd['Polarity']=(1-4*fd.theta/np.pi)
#daily index dfs['index']=((dfs.i+dfs.Polarity)/dfs.i)
fd['daily_index']=((fd.i+fd.Polarity)/fd.i)
```
